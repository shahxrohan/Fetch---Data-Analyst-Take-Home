### Check for NULL values in data

```python
# Replace empty (NaN) cells with 'NULL'
data = data.fillna('NULL')
```

```python
# Count the number of 'NULL' values in each column
null_counts = (data == 'NULL').sum()

# Plot the results
plt.figure(figsize=(10, 6))
bars = plt.bar(null_counts.index, null_counts.values)
plt.title('Count of NULL Values in Each Column', fontsize=16)
plt.xlabel('Columns', fontsize=14)
plt.ylabel('Count of NULL Values', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Add data labels on the bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             str(bar.get_height()), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

### Check and delete duplicate records

```python
# Count occurrences of each record, and group duplicate records
data = data.copy()
data['Occurrence_Count'] = data.groupby(data.columns.tolist()).transform('size')

# Drop duplicates to keep only unique records with their occurrence counts
data = data.drop_duplicates()

# Sorts records in descending order based on occurance
data = data.sort_values(by='Occurrence_Count', ascending=False)
```

```python
# View first 5 rows of data
data.head()
```

```python
# Keep only unique records, ensuring duplicates appear as a single entry
data = data.drop_duplicates(subset=data.columns[:-1], keep='first')
```

```python
# View first 5 rows of data
data.head()
```

```python
# Drop the 'Occurrence_Count' column
data = data.drop(columns=['Occurrence_Count'])
```
